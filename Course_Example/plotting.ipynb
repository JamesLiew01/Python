```python
#plot density
x=df['number'][:118]
y=df['density'][:118]
title='Density of Elements'
xlabel,ylabel='Elements',r'Density(g/$\rm cm^3$)'
fig, ax = plt.subplot(figsize=(3,3),dpi=100)
ax.set_title(title)
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
ax.plot(x,y)
```

```python
#plot density with y log bar
x=df['number'][:118]
y=df['density'][:118]
title='Density of Elements'
xlabel,ylabel='Elements',r'Density(g/$\rm cm^3$)'
fig, ax = plt.subplot(figsize=(3,3),dpi=100)
ax.set_title(title)
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
ax.set_yscale('log')
ax.plot(x,y)
```
