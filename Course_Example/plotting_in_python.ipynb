### **Multiple plots in one figure**

```python
from matplotlib import pyplot as plt
# generate some data
x = [0,1,2,3,4]
y = [0,2,4,6,8] 
# notice that ax is replaced with axes since we have multiple subplots
# it is now a list of ax objects, each for a subplot
num_subplots = 3
fig, axes = plt.subplots(1, num_subplots, figsize=(9,3), dpi=100)
#axes no longer an object
# plot some data
for i in range(num_subplots):
    axes[i].plot(x,y, label='2x', color='red', marker='o', markersize=10, linestyle='--', linewidth=2)
```

### **Axis labels, Title, and Legend**

### **Linestyle, Color, Marker, Transparency**

#### Linestyles

```python
num_rows = 3
num_columns = 3
size = 3
fig, axes = plt.subplots(num_rows, num_columns, figsize=(size * num_columns, size * num_rows), dpi=100)

# linestyles are used to change the style of the line
linestyles = ['-', '--', '-.', ':', 'solid', 'dashed', 'dashdot', 'dotted']
# plot some data
for i, linestyle in enumerate(linestyles):
    ax = axes[i//num_columns, i%num_columns]
    ax.plot(x, y, color='r', marker='o', # label removed
            markersize=10, linestyle=linestyle, linewidth=2)
    ax.set_title(f'linestyle={linestyle}')
```

#### Colors
In almost all places in matplotlib where a color can be specified by the user it can be provided as:

- an RGB or RGBA tuple of float values in [0, 1] (e.g., (0.1, 0.2, 0.5) or (0.1, 0.2, 0.5, 0.3))
- a hex RGB or RGBA string (e.g., '#0F0F0F' or '#0F0F0F0F')
- a string representation of a float value in [0, 1] inclusive for gray level (e.g., '0.5')
one of {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'}
- a X11/CSS4 color name
- a name from the xkcd color survey prefixed with 'xkcd:' (e.g., 'xkcd:sky blue')
- one of {'C0', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9'}
- one of {'tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'} which are the Tableau Colors from the T10 categorical palette (which is the default color cycle).

For the basic built-in colors, you can use a single letter

- b: blue
- g: green
- r: red
- c: cyan
- m: magenta
- y: yellow
- k: black
- w: white

```python
num_rows = 2
num_columns = 3
fig, axes = plt.subplots(num_rows, num_columns, figsize=(size * num_columns, size * num_rows), dpi=100)

# plot some data
# RGB Hex Codes, F is full, 0 is none, #FFFFFF is white, #000000 is black
colors = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#00FFFF', '#000000'] 
# You can also just use one letter color codes,
# colors = ['r', 'g', 'b', 'y', 'c', 'k']
for i, color in enumerate(colors):
    ax = axes[i//num_columns, i%num_columns]
    ax.plot(x, y, color=color, marker='o', # label removed
            markersize=10, linestyle='-', linewidth=2)
    ax.set_title(f'color={color}')
```

```python
num_rows = 4
num_columns = 4
size = 3
fig, axes = plt.subplots(num_rows, num_columns, figsize=(size * num_columns, size * num_rows), dpi=100)

# markers are used to highlight points on the graph
# you can use the following markers: '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
markers = ['+', 'o', '*', 's', 'x', 'D', 'd', 'p', 'H', 'h', 'v', '^', '<', '>']
# plot some data
for i, marker in enumerate(markers):
    ax = axes[i//num_columns, i%num_columns]
    ax.plot(x, y, color='r', marker=marker, # label removed
            markersize=10, linestyle='-', linewidth=2)
    ax.set_title(f'marker={marker}')
```

#### You can adjust the range of the x and y axis using xlim and ylim

```python
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis') 

# define the range of the axes
ax.set_xlim([-0.3, 8.3])
ax.set_ylim([-0.3, 8.3]) 
ax.plot(x,y, label='2x', color=colors[0], marker='o', 
        markersize=10, linestyle='-', linewidth=2)

# Add a legend
plt.legend()
```

### **Ticks**

#### Manually set the ticks

```python
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# define y ticks as prime numbers
ax.set_yticks([1, 2, 3, 5, 7])
# define x ticks as power of 2
ax.set_xticks([1, 2, 4])
ax.plot(x,y, label='2x', color='r', marker='o', 
        markersize=10, linestyle='-', linewidth=2)
# Add a legend
plt.legend()
```

#### different ticklocators

```python
num_rows = 2
num_columns = 3
size = 3
fig, axes = plt.subplots(num_rows, num_columns, figsize=(size * num_columns, size * num_rows), dpi=100)
# Add a title (specify font parameters with fontdict)
# plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# ticks
ticklocators = [plt.LinearLocator(numticks=3), plt.MultipleLocator(2), plt.MaxNLocator(3), plt.AutoLocator(), plt.LogLocator()]
ticklocator_names = ['LinearLocator', 'MultipleLocator', 'MaxNLocator', 'AutoLocator', 'LogLocator']
# plot some data
#MultipleLocator 2
#MaxN 3
for i, ticklocator in enumerate(ticklocators):
    ax = axes[i//num_columns, i%num_columns]
    ax.plot(x, y, color='r', marker='o', label='2x',
            markersize=10, linestyle='-', linewidth=2)
    ax.yaxis.set_major_locator(ticklocator)
    ax.set_ylim([0, 8])
    ax.set_title(f'Tick Locator: {ticklocator_names[i]}', fontsize=12)
    ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# plot some data
markers = ['o', '^', 's', 'D', 'x']
linestyles = ['-', '--', '-.', ':']
# RGB Hex Codes, F is full, 0 is none, #FFFFFF is white, #000000 is black
colors = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#00FFFF'] 
ax.set_xlim([0, 8])
ax.set_ylim([0, 8]) 
ax.yaxis.set_major_locator(plt.AutoLocator())
ax.plot(x,y, label='2x', color=colors[0], marker=markers[0], 
        markersize=10, linestyle=linestyles[0], linewidth=2)
# Add a legend
plt.legend()
```

### Heatmap

```python
# Heatmap
data = np.random.rand(3,3)
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.imshow(data)
```

#### cmaps

```python
# Heatmap
# cmap: str | Colormap = ..., norm: Normalize = ..., aspect: float | Literal['equal', 'auto'] = ..., 
# interpolation: str = ..., alpha: ArrayLike = ..., vmin: float = ..., vmax: float = ..., 
# origin: Literal['upper', 'lower'] = ..., extent: Sequence[float] = ..., *, 
# interpolation_stage: Literal['data', 'rgba'] = ..., filternorm: bool = True, filterrad: float = 4, 
# resample: bool = ..., url: str =
# heatmap
num_rows = 2
num_columns = 3
size = 3
fig, axes = plt.subplots(num_rows, num_columns, figsize=(size * num_columns, size * num_rows), dpi=100)
cmaps = ['viridis', 'plasma', 'inferno', 'magma', 'cividis']
# plot the histogram
for i, cmap in enumerate(cmaps):
    ax = axes[i//num_columns, i%num_columns]
    ax.imshow(data, cmap=cmap)
    ax.set_title(f'Cmap: {cmap}')
```

### Contour plot

```python
# coutour plot
data = np.random.rand(10,10)
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.contour(data)
```

#### Levels

```python
# coutour plot
num_rows = 2
num_columns = 3
size = 3
fig, axes = plt.subplots(num_rows, num_columns, figsize=(size * num_columns, size * num_rows), dpi=100)
levels = np.arange(0, 12, 2)
# plot the histogram
for i, level in enumerate(levels):
    ax = axes[i//num_columns, i%num_columns]
    ax.contour(data, levels=level)
    ax.set_title(f'Level: {level}')
```

### 3D plot

```python
# 3d plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure(figsize=(3,3), dpi=100)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
```
