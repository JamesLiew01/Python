# **Seaborn**
Similar to matplotlib, seaborn supports many different types of plots

## Different types of plots
Line plot, Scatter plot, Bar plot, Histogram, Box plot, Violin plot, Pie chart, Heatmap, Contour plot, 3D plot

### Line plot

```python
from matplotlib import pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.plot(x, y)
```

```python
# seaborn line plot
import seaborn as sns
plt.figure(figsize=(3,3), dpi=100)
sns.lineplot(x=x, y=y)
```

### scatter plot

```python
x = np.linspace(0, 3*np.pi, 100)
y = np.sin(x)
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.scatter(x, y)
```

```python
# seaborn scatter plot
plt.figure(figsize=(3,3), dpi=100)
sns.scatterplot(x=x, y=y)
```

### Bar plot

```python
x = np.linspace(0, 3*np.pi, 100)
y = np.sin(x)
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.bar(x, y)
```

```python
# seaborn bar plot
plt.figure(figsize=(3,3), dpi=100)
sns.barplot(x=x, y=y)
```

### Histogram

```python
# histogram
x = np.random.randn(1000)
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.hist(x, bins=30)
```

```python
# seaborn histogram
plt.figure(figsize=(3,3), dpi=100)
sns.histplot(x, bins=30)
```

```python
# seaborn kde plot
plt.figure(figsize=(3,3), dpi=100)
sns.kdeplot(x)
```

### Hist2d

```python
# matplotlib hist2d
x = np.random.randn(1000)
y = np.random.randn(1000)
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.hist2d(x, y, bins=30)
```

```python
# seaborn hist2d
plt.figure(figsize=(3,3), dpi=100)
sns.histplot(x=x, y=y, bins=30)
```

### Boxplot

```python
# matplotlib boxplot
data = np.random.randn(100)
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.boxplot(data)
```

```python
# Seaborn boxplot
plt.figure(figsize=(3,3), dpi=100)
sns.boxplot(data=data)
```

### Violin plot

```python
# Violin plot
x = np.random.randn(1000)
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.violinplot(x)
```

```python
# seaborn violin plot
plt.figure(figsize=(3,3), dpi=100)
sns.violinplot(x)
```

### Pie chart
Seaborn doesn't support pie chart, but we can use matplotlib to plot pie chart

```python
# Pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [25, 25, 25, 25]
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.pie(sizes, labels=labels)
```

### Heatmap

```python
# Heatmap
data = np.random.rand(3,3)
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.imshow(data)
```

```python
# seaborn heatmap
plt.figure(figsize=(3,3), dpi=100)
sns.heatmap(data)
```

### Contour plot
Seaborn doesn't support contour plot, but we can use matplotlib to plot contour plot

```python
# coutour plot
data = np.random.rand(10,10)
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.contour(data)
```

### 3D plot
SeaBorn doesn't support 3D plot, but we can use matplotlib to plot 3D plot

```python
# 3d plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure(figsize=(3,3), dpi=100)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)
```

```python
# seaborn pairplot
import seaborn as sns
iris = sns.load_dataset('iris')
sns.pairplot(iris)
```

```python
# seaborn facetplot
tips = sns.load_dataset('tips')
sns.FacetGrid(tips, col='time', row='smoker').map(sns.scatterplot, 'total_bill', 'tip')
```

```python
# seaborn jointgrid
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
```

```python
# seaborn clustermap
data = np.random.rand(10,10)
sns.clustermap(data)
```

```python
# seaborn pairgrid
g = sns.PairGrid(iris)
g.map(sns.scatterplot)
```

```python
# seaborn lmplot
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# seaborn jointplot
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
# seaborn relplot
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
# seaborn catplot
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
# seaborn swarmplot
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
# seaborn stripplot
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
# seaborn pointplot
sns.pointplot(x='day', y='total_bill', data=tips)
```

```python
# seaborn countplot
sns.countplot(x='day', data=tips)
```

```python
# seaborn boxenplot
sns.boxenplot(x='day', y='total_bill', data=tips)
```

```python
# seaborn relplot
sns.relplot(x='day', y='total_bill', data=tips)
```
