#### Let's separate elements by their state.
It is not surprising that gaseous elements have low density.  <br>
It turns out that the density for gas is in the unit of g/L (ideal gas), while the density for solid and liquid is in the unit of g/cm^3. Let's convert the density of gas to g/cm^3.  <br>

```python
# Define variables
title = 'Density of Elements'
xlabel, ylabel = 'Atomic number', r'Density (g/$\rm cm^3$)'
# Create fig and ax
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
# Add elements into ax
ax.set_title(title)
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
ax.set_yscale('log')
# Plot gas
y = df[df['boil'] < 298.15]['density'] / 1000
x = df[df['boil'] < 298.15]['number']
ax.scatter(x, y, label='gas', edgecolor='k')
# Plot liquid
y = df[(df['boil'] > 298.15) & (df['melt'] < 298.15)]['density']
x = df[(df['boil'] > 298.15) & (df['melt'] < 298.15)]['number']
ax.scatter(x, y, color='r', label='liquid', edgecolor='k')
# Plot solid
y = df[(df['boil'] > 298.15) & (df['melt'] > 298.15)]['density']
x = df[(df['boil'] > 298.15) & (df['melt'] > 298.15)]['number']
ax.scatter(x, y, color='g', label='solid', edgecolor='k')

ticklocator = plt.MultipleLocator(30)
ax.xaxis.set_major_locator(ticklocator)
ticklocator = plt.MultipleLocator(15)
ax.xaxis.set_minor_locator(ticklocator)
ax.legend()
```

It is clear that density is linearly dependent on atomic number (atomic mass) for gaseous elements. <br>
Let's use what we learned about scipy to fit a linear line

```python
from scipy.optimize import curve_fit

# Define a linear function
def linear(x, m, c):
    return m*x + c
# Obtain data for gaseous elements
y = df[df['boil'] < 298.15]['density'] / 1000
x = df[df['boil'] < 298.15]['number']
# Fit
popt, cov = curve_fit(linear, x, y) 
# Fit line
x_fit = np.linspace(0, 118, 100) #將0-118 分成100個點
y_fit = linear(x_fit, *popt) #算出trendline的y值
```

Plot the fit line

```python
# Plot density
title = 'Density of Elements'
xlabel, ylabel = 'Atomic number', r'Density (g/$\rm cm^3$)'
fig, ax = plt.subplots(figsize=(3,3), dpi=100)
ax.set_title(title)
ax.set_xlabel(xlabel)
ax.set_ylabel(ylabel)
# ax.set_yscale('log')
ax.scatter(x, y, label='gas', edgecolor='k')
ax.plot(x_fit, y_fit, 'k--', label='fit')
ax.legend()
```
